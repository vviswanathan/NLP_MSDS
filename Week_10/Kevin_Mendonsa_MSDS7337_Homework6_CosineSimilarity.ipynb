```python
len(Titles)
```

```python
# Create the Document Term Matrix
count_vectorizer = CountVectorizer(stop_words='english')
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(Titles)
```

```python
doc_term_matrix = sparse_matrix.todense()
df_Titles_Similarity = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names(), 
                  index=[   'Title1',
                            'Title2',
                            'Title3',
                            'Title4',
                            'Title5',
                            'Title6',
                            'Title7',
                            'Title8',
                            'Title9',
                            'Title10',
                            'Title11',
                            'Title12',
                            'Title13',
                            'Title14',
                            'Title15',
                            'Title16',
                            'Title17',
                            'Title18',
                            'Title19',
                            'Title20',
                            'Title21',
                            'Title22',
                            'Title23',
                            'Title24'
                            ])
df_Titles_Similarity 
```

```python
cosine_similarity_score = cosine_similarity(df_Titles_Similarity , df_Titles_Similarity )[0]

cosine_similarity_score
```

```python
# Which two titles are the most similar to each other? 
# Where do they rank, among the first 24 results?
print("Title 13: ", Titles[12])
print("Title 16: ", Titles[15])

# Which are the most dissimilar? 
print("Title 1: ", Titles[0])
print("Title 23: ", Titles[22])
```
