## Answer
Edit distance between two strings is calculated by using Levenshtein distance. The distance between the strings are dependent on the number of edits that are needed to match the words ie. insertion, deletion and substitution. I first manually calculated the distance using the following concept:
1. If the characters at a particular position is same, go with the diagonal value
2. If the characters are different, find the minimum value to the left, diagonal and top and then add one.

## Reference to the image below that is hand drawn

1. We have two strings "trial" and "trail". The edit distance calculated manually is 2 as indicated in the image
2. Let us review the value for the character a. As per the diagram it is "1". How did we arrive at the value 1? Since both the characters are same, we go with the diagonal value ie. 1. Please refer to the cell "r" and "i".
3. Now let us review the value for the character "l" and "a" which is 2. How did we arrive at the value 2? The values to the left is "1", diagnol is "2" and above is "3". The minimum is 1 and add 1 and that makes 2.

## Formula (consider strings S1 and S2)
If S1[i] = S2[j], consider the diagonal value
 T[i][j] = T[i-1][j-1]
Else
T[i][j] = min(T[i-1][j], T[i-1][j-1],T[i][j-1]) +1

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return content
```

```python
content_without_stop_words = content_fraction(tokens)
```
